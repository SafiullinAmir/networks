### Task 1

Compute degree centrality, Pagerank and  HubAuthorities scores for the [flickr](https://www.hse.ru/data/2016/02/24/1139618969/flickr.mat) network. 

Data contains sparse matrix A and list of user names.
This is a “denser” part of the Flickr photo sharing site friendship graph from 2006. Edge direction corresponds to friendship requests (following). Some of the links are reciprocal,others not.  

Provide top 50 names in each ranking, compare results

You can load .mat files with the following commands:

```python
import scipy.io
import numpy as np
import networkx as nx
import pandas as pd

data = scipy.io.loadmat('flickr.mat')
```

Load data to networkx DiGraph

```python
G = nx.from_scipy_sparse_matrix(data['A'], create_using=nx.DiGraph())
```

Calculate in and out degree centrality, pagerank and hubs and authorities

```python
idc = nx.in_degree_centrality(G)
odc = nx.out_degree_centrality(G)
pr = nx.pagerank_scipy(G)
hubs, authorities = nx.hits_scipy(G)
```

Sort and store names for top50 users in each rank

```python
ranknames = ['In degree centrality', 'Out degree centrality', 'PageRank', 'Hub', 'Authority']
ranklist = [idc, odc, pr, hubs, authorities]

df = pd.DataFrame(index=range(1, 51))

for i in range(5):
    name = ranknames[i]
    rank = ranklist[i]
    sorted_rank = sorted(rank.items(), key=lambda x: x[1], reverse=True)
    top_50 = [data['names'][i[0]].rstrip() for i in sorted_rank[:50]]
    df.loc[:, name] = top_50
```

```python
df
```

Compare rankings

```python
from scipy.stats import spearmanr

comp = pd.DataFrame(index=[ranknames[i]+' and '+ranknames[j] for i in range(5) for j in range(i+1, 5)])

spearmans = []
pvalues = []
intersects = []

for i in range(5):
    for j in range(i+1, 5):
        s, p = spearmanr(df.loc[:, ranknames[i]].values, df.loc[:, ranknames[j]].values)
        spearmans.append(s)
        pvalues.append(p)
        intersects.append(np.intersect1d(df.loc[:, ranknames[i]].values, 
                                         df.loc[:, ranknames[j]].values).shape[0])
        
comp.loc[:, 'Spearman rank coeff'] = spearmans
comp.loc[:, 'Pvalues for spearman'] = pvalues
comp.loc[:, 'Intersections'] = intersects

comp
```

Our calculations indicate that in degree centrality and the PageRank produce rankings that match each closer than other methods.

Here are the [Facebook friendship graphs](https://www.hse.ru/data/2016/03/05/1125656999/universities.zip) from several US universities from 2005 (one year after fb launch).

Data contains a A matrix (sparse) and a "local_info" variable, one row per node: 
a student/faculty status flag, gender, major, second major/minor (if applicable), dorm/house, year, and high school. 
Missing data is coded 0.

Compute node degree assortativity (mixining by node degree) and assortativity coefficient (modularity) for gender, major, dormitory, year, high school for all universities and compare the results.

```python
berkeley_data = scipy.io.loadmat('universities/Berkeley13.mat')
caltech_data = scipy.io.loadmat('universities/Caltech36.mat')
harvard_data = scipy.io.loadmat('universities/Harvard1.mat')
oklahoma_data = scipy.io.loadmat('universities/oklahoma97.mat')
princeton_data = scipy.io.loadmat('universities/Princeton12.mat')
texas_data = scipy.io.loadmat('universities/texas80.mat')
```

```python
def read_graph(data):
    graph = nx.from_scipy_sparse_matrix(data['A'], create_using=nx.DiGraph())
    for i in range(len(graph)):
        for j in range(7):
            graph.node[i][j] = data['local_info'][i, j]
    return graph

berkeley_graph = read_graph(berkeley_data)
caltech_graph = read_graph(caltech_data)
harvard_graph = read_graph(harvard_data)
oklahoma_graph = read_graph(oklahoma_data)
princeton_graph = read_graph(princeton_data)
texas_graph = read_graph(texas_data)
```

```python
coefficient_names = ['Status', 'Gender', 'Major', '2nd major', 
                     'Dorm', 'Year', 'High School', 'Degree Assortativity']
uni_names = ['Berkeley', 'CalTech', 'Harvard', 'Oklahoma', 'Princeton', 'Texas']
unis = [berkeley_graph, caltech_graph, harvard_graph,
        oklahoma_graph, princeton_graph, texas_graph]

coefs = pd.DataFrame(index=coefficient_names, columns=uni_names)

for i in range(6):
    cs = []
    for j in range(7):
        cs.append(nx.attribute_assortativity_coefficient(unis[i], j))
    cs.append(nx.degree_assortativity_coefficient(unis[i]))
    coefs.loc[:, uni_names[i]] = cs
```
