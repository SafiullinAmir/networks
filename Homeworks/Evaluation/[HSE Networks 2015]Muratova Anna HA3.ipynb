### Task 1

Compute degree centrality, Pagerank and  HubAuthorities scores for the [flickr](https://snap.stanford.edu/data/web-flickr.html) network. 

Data contains sparse matrix A and list of user names.
This is a “denser” part of the Flickr photo sharing site friendship graph from 2006. Edge direction corresponds to friendship requests (following). Some of the links are reciprocal,others not.  

Provide top 50 names in each ranking, compare results

You can load .mat files with the following commands:

```python
import scipy.io

data = scipy.io.loadmat('filename.mat')
```

### <hr /> Code of program

```python
import scipy.io
import networkx as nx
#import time
  
print "============================================="
print "    HA-3    Task-1    Muratova Anna"
print "============================================="
print "Compute degree centrality, Pagerank and HubAuthorities scores for the flickr network"
print "============================================="

data = scipy.io.loadmat('flickr.mat')
print "Data Structure:"
print scipy.io.whosmat('flickr.mat')
print "============================================="
A = data['A']
#names = data['names']

G = nx.from_scipy_sparse_matrix(A)
"""
The degree centrality for a node is the fraction of nodes it is connected to.
Returns: Dictionary of nodes with degree centrality as the value.
"""
centrality = nx.degree_centrality(G)

print "Centrality (top 50):"
print "===================="
centrality_sorted = sorted(centrality.items(), reverse=True, key=lambda (k,v): (v,k))
for i in xrange(50):
    print i+1, data['names'][centrality_sorted[i][0]], centrality_sorted[i]
print "============================================="

page_rank = nx.pagerank(G) 
"""
PageRank is a ranking of the nodes in the graph 
         based on the structure of the incoming links
Returns: Dictionary of nodes with PageRank as value
"""
print "PageRank (top 50):"
print "=================="
page_rank_sorted = sorted(page_rank.items(), reverse=True, key=lambda (k,v): (v,k))
for i in xrange(50):
    print i+1, data['names'][page_rank_sorted[i][0]], page_rank_sorted[i]
print "============================================="

hubs, authorities = nx.hits(G)
"""
The HITS algorithm computes two numbers for node. 
Authorities estimates the node value based on the incoming links. 
Hubs estimates the node value based on outgoing links.
Returns: hubs and authorities - two dictionaries keyed by node 
                                containing the hub and authority values.
"""
print "Hubs (top 50):"
print "=============="
hubs_sorted = sorted(page_rank.items(), reverse=True, key=lambda (k,v): (v,k))
for i in xrange(50):
    print i+1, data['names'][hubs_sorted[i][0]], hubs_sorted[i]
print "============================================="

print "Authorities (top 50):"
print "====================="
authorities_sorted = sorted(authorities.items(), reverse=True, key=lambda (k,v): (v,k))
for i in xrange(50):
    print i+1, data['names'][authorities_sorted[i][0]], authorities_sorted[i]
print "============================================="
```

### <hr /> Task 2

Here are the [Facebook friendship graphs](https://snap.stanford.edu/data/egonets-Facebook.html) from several US universities from 2005 (one year after fb launch).

Data contains a A matrix (sparse) and a "local_info" variable, one row per node: 
a student/faculty status flag, gender, major, second major/minor (if applicable), dorm/house, year, and high school. 
Missing data is coded 0.

Compute node degree assortativity (mixining by node degree) and assortativity coefficient (modularity) for gender, major, dormitory, year, high school for all universities and compare the results.

### <hr /> Code of program

```python
import scipy.io
import networkx as nx
#import time
def assortativity(u, data):    # Assortativities calculaton

    A = data['A']
    attr_values = data['local_info']
    attr_names = ['status', 'gender', 'major', 'second', 'dorm/house', 'year', 'high school']
    print "============================================="
    
    G = nx.from_scipy_sparse_matrix(A)

    for i in xrange(len(G)):        # set node attributes in the graph
        for j in xrange(len(attr_names)):
            G.node[i][attr_names[j]] = attr_values[i][j]   

    print u, "Degree assortativity of graph =", nx.degree_assortativity_coefficient(G)
    print "---------------------------------------------"

    print u, "Assortativity coefficients:"
    print "---------------------------"
    for i in xrange(len(attr_names)):
        print attr_names[i].ljust(12), nx.attribute_assortativity_coefficient(G, attr_names[i])


print "============================================="
print "    HA-3    Task-2    Muratova Anna"
print "============================================="

universities = ['Caltech36', 'Berkeley13', 'Harvard1', 'Oklahoma97', 'Princeton12', 'Texas80']

for u in universities:
    data = scipy.io.loadmat('universities\\' + u)
    print u, "Data Structure:"
    print scipy.io.whosmat('universities\\' + u)          
    assortativity(u, data)
    print "============================================="
```
