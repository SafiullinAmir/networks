### Task 1

Compute degree centrality, Pagerank and  HubAuthorities scores for the [flickr](https://snap.stanford.edu/data/web-flickr.html) network. 

Data contains sparse matrix A and list of user names.
This is a “denser” part of the Flickr photo sharing site friendship graph from 2006. Edge direction corresponds to friendship requests (following). Some of the links are reciprocal,others not.  

Provide top 50 names in each ranking, compare results

You can load .mat files with the following commands:

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import pandas as pd
%matplotlib inline
```

```python
flickr_data = scipy.io.loadmat('./flickr.mat')
edges = flickr_data['A']
names = flickr_data['names']
flickr_graph = nx.DiGraph(edges)
```

```python
# посчитаем degree centrality, выделим top 50 
dc = nx.degree_centrality(flickr_graph)
sorted_dc = sorted(dc, key = dc.get, reverse = True)
sorted_dc_top50 = sorted_dc[:50]


# посчитаем PageRank
pr = nx.pagerank(flickr_graph)
sorted_pr = sorted(pr, key = pr.get, reverse = True)
sorted_pr_top50 = sorted_pr[:50]


# HubAuthorities 
hub, auth = nx.hits(flickr_graph)
sorted_hub = sorted(hub, key = hub.get, reverse =True)
sorted_auth = sorted(auth, key = auth.get, reverse =True)

sorted_hub_top50 = sorted_hub[:50]
sorted_auth_top50 = sorted_auth[:50]
```

```python
# для сравнения выведим табличку
ranks = pd.DataFrame(columns = ['Degree Centrality', 'Page Rank', 'Hubs', 'Authorities'])
for i in range(0,50):
    ranks.loc[i] = [names[sorted_dc_top50[i]], names[sorted_pr_top50[i]], names[sorted_hub_top50[i]], names[sorted_auth_top50[i]]]
ranks
```

### Task 2

Here are the [Facebook friendship graphs](https://snap.stanford.edu/data/egonets-Facebook.html) from several US universities from 2005 (one year after fb launch).

Data contains a A matrix (sparse) and a "local_info" variable, one row per node: 
a student/faculty status flag, gender, major, second major/minor (if applicable), dorm/house, year, and high school. 
Missing data is coded 0.

Compute node degree assortativity (mixining by node degree) and assortativity coefficient (modularity) for gender, major, dormitory, year, high school for all universities and compare the results.

```python
def load_mat(string_mat):
    data_mat = scipy.io.loadmat(string_mat)
    edges = data_mat['A']
    features = data_mat['local_info']
    graph = nx.Graph(data = edges)
    
    gender_dict = {i: value for i, value in enumerate(features[:,1])}
    major_dict = {i: value for i, value in enumerate(features[:,2])}
    dormitory_dict = {i: value for i, value in enumerate(features[:,4])}
    year_dict = {i: value for i, value in enumerate(features[:,5])}
    highschool_dict = {i: value for i, value in enumerate(features[:,6])}

    nx.set_node_attributes(graph,'gender', gender_dict)
    nx.set_node_attributes(graph,'major', major_dict)
    nx.set_node_attributes(graph,'dormitory', dormitory_dict)
    nx.set_node_attributes(graph,'year', year_dict)
    nx.set_node_attributes(graph,'highschool', highschool_dict)
    
    return graph 


def assort_coef(graph):
    r = nx.degree_assortativity_coefficient(graph)
    gender_r = nx.attribute_assortativity_coefficient(graph,'gender')
    major_r = nx.attribute_assortativity_coefficient(graph,'major')
    dormitory_r = nx.attribute_assortativity_coefficient(graph,'dormitory')
    year_r = nx.attribute_assortativity_coefficient(graph,'year')
    highschool_r = nx.attribute_assortativity_coefficient(graph,'highschool')
    
    print '''node degree assortativity = {0}, 
    assortativity coefficient (modularity) for gender = {1}, 
    assortativity coefficient (modularity) for major = {2}, 
    assortativity coefficient (modularity) for dormitory = {3}, 
    assortativity coefficient (modularity) for year = {4}, 
    assortativity coefficient (modularity) for high school = {5}'''.format(r, gender_r, major_r, dormitory_r, 
                                                                           year_r, highschool_r)
```

```python
graph_Berk = load_mat('./universities/Berkeley13.mat')
print 'ok!'
graph_Calt = load_mat('./universities/Caltech36.mat')
print 'ok!'
graph_Harv = load_mat('./universities/Harvard1.mat')
print 'ok!'
graph_Oklah = load_mat('./universities/Oklahoma97.mat')
print 'ok!'
graph_Princ = load_mat('./universities/Princeton12.mat')
print 'ok!'
graph_Tex = load_mat('./universities/Texas80.mat')
```

```python
assort_coef(graph_Berk)
```
