### Task 1

The following function constructs a graph with $N$ nodes according to model A and returns except the graph the vector of time points, the vectors of average shortest path values, average clustering coefficient values and average node degree values and at this time points. The graph starts to grow from $m$ unconnected nodes and at every time-step a new node is added and is connected to m randomly selected existing nodes.

```python
def modelA(m, N):
    n0 = m
    path = []
    clust = []
    deg = []
    kk = []
    G = nx.Graph();
    G.add_nodes_from(range(n0))
    for k in range(n0, N):
        G.add_node(k)
        chosen_nodes = range(k)
        np.random.shuffle(chosen_nodes)
        chosen_nodes = chosen_nodes[:m]
        for ch_n in chosen_nodes:
            G.add_edge(k,ch_n)
        kk.append(k)
        path.append(nx.average_shortest_path_length(G))
        clust.append(nx.average_clustering(G))
        deg.append(np.mean(G.degree().values()))
    return G, kk, path, clust, deg
```

The next function constructs a graph with $N$ nodes according to model B and returns except the graph the vector of time points, The vectors of average shortest path values, average clustering coefficient values and average node degree values and at this time points. At the first time-step one node is connected with $m$ other nodes. At the next time_steps nodes for connection are seleted according to preferential attachment. Parameter $lam$ is added to the model to make nodes with zero degree also probable to be selected. One node can not be selected twice. Also, the graph constructed under this model is not connected frequently, so average shortest path is not always available.

```python
def modelB(m, N):
    kk = []
    path = []
    clust = []
    deg = []
    G = nx.Graph()
    nodes = range(N)
    G.add_nodes_from(nodes)
    lam = 0.001
    G.add_edges_from([(m,x) for x in range(m)])
    not_chosen_nodes = range(m+1,N)
    if not nx.is_connected(G):
        path.append(float('Inf'))
    else:
        path.append(nx.average_shortest_path_length(G))
    clust.append(nx.average_clustering(G))
    kk.append(m)
    deg.append(np.mean(G.degree().values()))
    for k in range(m+1,N):
        ch_node = not_chosen_nodes[np.random.randint(len(not_chosen_nodes))]
        not_chosen_nodes.remove(ch_node)
        neighbors = G.neighbors(ch_node)
        non_neighbors = nodes[:]
        for nn in neighbors:
            non_neighbors.remove(nn)
        non_neighbors.remove(ch_node)
        len_non = len(non_neighbors)
        if len_non <= m:
            for nn in non_neighbors:
                G.add_edge(ch_node, nn)
        else:
            degrees = G.degree(non_neighbors)
            sum_deg = sum(degrees.values())+lam*len_non
            cur_sum = [0]
            for nn in non_neighbors:
                cur_sum.append(cur_sum[-1] + degrees[nn]+lam)
            idx = set([])
            while len(idx) < m:
                p = sum_deg*np.random.rand()
                i = 0;
                while cur_sum[i] < p:
                    i = i+1
                idx.add(i-1)
            idx = list(idx)
            for i in idx:
                G.add_edge(ch_node, non_neighbors[i])
        if not nx.is_connected(G):
            path.append(float('Inf'))
        else:
            path.append(nx.average_shortest_path_length(G))
        clust.append(nx.average_clustering(G))
        kk.append(k)
        deg.append(np.mean(G.degree().values()))
    return G, kk, path, clust, deg
        
```

The functions *pdf* and *cdf* are used to calculate pdf and cdf correspondingly in both tasks.

```python
def pdf(G):
    degrees = G.degree().values()
    max_deg = max(degrees)
    pdfv = []
    kk = range(1,max_deg+1)
    for k in kk:
        pdfv.append(sum([x == k for x in degrees]))
    pdfv = [x*1./len(degrees) for x in pdfv]
    return kk, pdfv
```

```python
def cdf(G):
    degrees = G.degree().values()
    max_deg = max(degrees)
    cdfv = []
    kk = range(1,max_deg+1)
    for k in kk:
        cdfv.append(sum([x<=k for x in degrees]))
    cdfv = [x*1./len(degrees) for x in cdfv]
    return kk, cdfv
```

#### The case of undirected graph

```python
def uVCM(p,n0,q,N):
    step = N/10
    deg = []
    path = []
    clust = []
    kk = []
    G = nx.fast_gnp_random_graph(n0,p)
    while not nx.is_connected(G):
        G = nx.fast_gnp_random_graph(n0,p)
    kk.append(n0-1)
    path.append(nx.average_shortest_path_length(G))
    clust.append(nx.average_clustering(G))
    deg.append(np.mean(G.degree().values()))
    for k in range(n0,N):
        gorder = G.order()
        ch_node = np.random.randint(gorder)
        neighbors = G.neighbors(ch_node)
        not_neighbors = G.nodes()[:]
        for n in neighbors:
            not_neighbors.remove(n)
        G.add_node(gorder)
        for n in neighbors:
            if np.random.rand() < q:
                G.add_edge(gorder,n)
            else:
                if not_neighbors:
                    nn = not_neighbors[np.random.randint(len(not_neighbors))]
                    G.add_edge(gorder,nn)
                    not_neighbors.remove(nn)
                else:
                    G.add_edge(gorder,n)
        if k%step == 0:
            kk.append(k)
            path.append(nx.average_shortest_path_length(G))
            clust.append(nx.average_clustering(G))
            deg.append(np.mean(G.degree().values()))
    return G, kk, path, clust,deg
```

We start from the random graph on 5 nodes with 0.3 probability of connection. Let us construct the graph under the given model on 20 nodes with parameter $q$ equals to 0.5 and draw it to be sure that the corresponding fiction works correctly.

```python
p = 0.3
n0 = 5
q = 0.5
N = 20
G, kk, path, clust, deg = uVCM(p,n0,q,N)
nx.draw_networkx(G)
```

#### Directed version

```python
def dVCM(p,n0,q,N):
    step = N/10
    path = []
    kk = []
    in_deg = []
    out_deg = []
    G = nx.fast_gnp_random_graph(n0,p,directed = True)
    while not nx.is_semiconnected(G):
        G = nx.fast_gnp_random_graph(n0,p,directed = True)
    kk.append(n0-1)
    path.append(nx.average_shortest_path_length(G))
    in_deg.append(np.mean(G.in_degree().values()))
    out_deg.append(np.mean(G.out_degree().values()))
    for k in range(n0,N):
        gorder = G.order()
        ch_node = np.random.randint(gorder)
        successors = G.successors(ch_node)
        not_successors = G.nodes()[:]
        for n in successors:
            not_successors.remove(n)
        G.add_node(gorder)
        for n in successors:
            if np.random.rand() < q:
                G.add_edge(gorder,n)
            else:
                if not_successors:
                    nn = not_successors[np.random.randint(len(not_successors))]
                    G.add_edge(gorder,nn)
                    not_successors.remove(nn)
                else:
                    G.add_edge(gorder,n)
        predecessors = G.predecessors(ch_node)
        not_predecessors = G.nodes()[:]
        for n in predecessors:
            not_predecessors.remove(n)
        G.add_node(gorder)
        for n in predecessors:
            if np.random.rand() < q:
                G.add_edge(n,gorder)
            else:
                if not_predecessors:
                    nn = not_predecessors[np.random.randint(len(not_predecessors))]
                    G.add_edge(nn,gorder)
                    not_predecessors.remove(nn)
                else:
                    G.add_edge(n,gorder)
        if k % step == 0:
            kk.append(k)
            path.append(nx.average_shortest_path_length(G))
            in_deg.append(np.mean(G.in_degree().values()))
            out_deg.append(np.mean(G.out_degree().values()))
    return G, kk, path, in_deg, out_deg
```

Clustering coefficient for directed graph is not defined. It can be calculated for the correspnding undirected graph, but this case was considered above. Except clustering let us calculate in_degree and out_degree separately.

```python
def in_pdf(G):
    degrees = G.in_degree().values()
    max_deg = max(degrees)
    pdfv = []
    kk = range(1,max_deg+1)
    for k in kk:
        pdfv.append(sum([x == k for x in degrees]))
    pdfv = [x*1./len(degrees) for x in pdfv]
    return kk, pdfv
```

```python
def out_pdf(G):
    degrees = G.out_degree().values()
    max_deg = max(degrees)
    pdfv = []
    kk = range(1,max_deg+1)
    for k in kk:
        pdfv.append(sum([x == k for x in degrees]))
    pdfv = [x*1./len(degrees) for x in pdfv]
    return kk, pdfv
```

```python
def in_cdf(G):
    degrees = G.in_degree().values()
    max_deg = max(degrees)
    cdfv = []
    kk = range(1,max_deg+1)
    for k in kk:
        cdfv.append(sum([x<=k for x in degrees]))
    cdfv = [x*1./len(degrees) for x in cdfv]
    return kk, cdfv
```

```python
def out_cdf(G):
    degrees = G.out_degree().values()
    max_deg = max(degrees)
    cdfv = []
    kk = range(1,max_deg+1)
    for k in kk:
        cdfv.append(sum([x<=k for x in degrees]))
    cdfv = [x*1./len(degrees) for x in cdfv]
    return kk, cdfv
```
