### Task 1

Compute degree centrality, Pagerank and  HubAuthorities scores for the [flickr](https://snap.stanford.edu/data/web-flickr.html) network. 

Data contains sparse matrix A and list of user names.
This is a “denser” part of the Flickr photo sharing site friendship graph from 2006. Edge direction corresponds to friendship requests (following). Some of the links are reciprocal,others not.  

Provide top 50 names in each ranking, compare results

You can load .mat files with the following commands:

```python
import scipy.io
data = scipy.io.loadmat('flickr.mat')
```

**Let us see how data looks like:**

```python
matrix = data.values()[0]
names = data.values()[2]
print scipy.shape(matrix), scipy.shape(names)
```

**Then we load the data to the graph object and construct a data frame indexed with names of nodes:**

```python
import networkx as nx
G = nx.from_scipy_sparse_matrix(matrix, create_using=nx.DiGraph())
```

```python
import pandas as pd
df = pd.DataFrame(matrix.todense(),index=names)
df.head()
```

**Computing PageRank values for all the nodes, adding them to a data frame, then sorting by descending value of PageRank:**

```python
pagerank = nx.pagerank(G).values()
```

```python
df["pagerank"] = pagerank
```

```python
df.head()
```

```python
top_result = df.sort_values(by="pagerank", ascending = False)
top_result.head(20)
```

**Forming the list of top 50:**

```python
pageRankTop = list(top_result.index[:50])
```

**Computing HITS values, adding a column of these values to our dataframe, providing top 50:**

```python
hits = nx.hits(G)
```

```python
hubs = pd.DataFrame.from_dict(hits[0].items())
authorities = pd.DataFrame.from_dict(hits[1].items())
```

```python
hubs.columns = ['node','hubs']
authorities.columns = ['node','authorities']
```

```python
df['hubs'] = list(hubs['hubs'])
df['authorities'] = list(authorities['authorities'])
```

```python
df.head()
```

```python
top_result_hubs = df.sort_values(by="hubs", ascending = False)
top_result_hubs.head(20)
```

```python
hubsTop = list(top_result_hubs.index[:50])
```

```python
top_result_authorities = df.sort_values(by="authorities", ascending = False)
```

```python
authoritiesTop = list(top_result_authorities.index[:50])
```

**Finally forming the table comparing two ratings' results:**

```python
rating = pd.DataFrame(pageRankTop, columns = ['pageRankTop'])
rating['hubsTop'] = hubsTop
rating['authoritiesTop'] = authoritiesTop
```

```python
rating
```

**As we see, the first place in PageRank and Authorities ratings is the same node, it shows the power of this user in terms of incoming links (both algorithms are based on power of in-degree), all the rest of the nodes from PageRank rating mostly appear in the 'Authorities' rating, sometimes even get the same place, but much more often they change its position. 'Hubs' rating is more different because it shows the power of out-degree of the node **

### <hr /> Task 2

Here are the [Facebook friendship graphs](https://snap.stanford.edu/data/egonets-Facebook.html) from several US universities from 2005 (one year after fb launch).

Data contains a A matrix (sparse) and a "local_info" variable, one row per node: 
a student/faculty status flag, gender, major, second major/minor (if applicable), dorm/house, year, and high school. 
Missing data is coded 0.

Compute node degree assortativity (mixining by node degree) and assortativity coefficient (modularity) for gender, major, dormitory, year, high school for all universities and compare the results.

**Downloading the data: we need only 3 first items in .mat files:**

```python
import pandas as pd
import networkx as nx
import scipy.io

berkeley = scipy.io.loadmat('Berkeley13.mat').values()[:2]
caltech = scipy.io.loadmat('Caltech36.mat').values()[:2]
harvard = scipy.io.loadmat('Harvard1.mat').values()[:2]
oklahoma = scipy.io.loadmat('Oklahoma97.mat').values()[:2]
princeton = scipy.io.loadmat('Princeton12.mat').values()[:2]
texas = scipy.io.loadmat('Texas80.mat').values()[:2]

universities = (berkeley, caltech, harvard, oklahoma, princeton, texas)
```

```python
names = ('berkeley', 'caltech', 'harvard', 'oklahoma', 'princeton', 'texas')
universities = dict(zip(names, universities))
```

**This is how one data item looks like:**

```python
universities.items()[0]
```

**Retrieving local information:**

```python
harvard_local = pd.DataFrame(universities.items()[0][1][1])
harvard_local.columns = ['student/faculty status flag', 'gender', 'major', 'second major/minor', 'dorm/house', 
                         'year', 'high school']
harvard_local.head(5)
```

```python
harvard_local.groupby("gender").sum()
```

```python
princeton_local = pd.DataFrame(universities.items()[1][1][1])
princeton_local.columns = ['student/faculty status flag', 'gender', 'major', 'second major/minor', 'dorm/house', 
                         'year', 'high school']

oklahoma_local = pd.DataFrame(universities.items()[2][1][1])
oklahoma_local.columns = ['student/faculty status flag', 'gender', 'major', 'second major/minor', 'dorm/house', 
                         'year', 'high school']

caltech_local = pd.DataFrame(universities.items()[3][1][1])
caltech_local.columns = ['student/faculty status flag', 'gender', 'major', 'second major/minor', 'dorm/house', 
                         'year', 'high school']

berkeley_local = pd.DataFrame(universities.items()[4][1][1])
berkeley_local.columns = ['student/faculty status flag', 'gender', 'major', 'second major/minor', 'dorm/house', 
                         'year', 'high school']

texas_local = pd.DataFrame(universities.items()[5][1][1])
texas_local.columns = ['student/faculty status flag', 'gender', 'major', 'second major/minor', 'dorm/house', 
                         'year', 'high school']
```

**Constructing graphs from sparse matrices:**

```python
bG = nx.from_scipy_sparse_matrix(universities.items()[4][1][0])
cG = nx.from_scipy_sparse_matrix(universities.items()[3][1][0])
hG = nx.from_scipy_sparse_matrix(universities.items()[0][1][0])
oG = nx.from_scipy_sparse_matrix(universities.items()[2][1][0]) 
pG = nx.from_scipy_sparse_matrix(universities.items()[1][1][0]) 
tG = nx.from_scipy_sparse_matrix(universities.items()[5][1][0])
```

**Computing degree assortativity coefficients of different social networks:**

```python
degree_assort_coefs = dict(zip(names, [nx.degree_assortativity_coefficient(bG),nx.degree_assortativity_coefficient(cG),
                                nx.degree_assortativity_coefficient(hG),nx.degree_assortativity_coefficient(oG),
                                nx.degree_assortativity_coefficient(pG),nx.degree_assortativity_coefficient(tG)]))
```

```python
for key, val in degree_assort_coefs.items():
    print key, val
```

**Let's set the node attributes for all the graphs:**

```python
dictb = berkeley_local.to_dict()
dictp = princeton_local.to_dict()
dicto = oklahoma_local.to_dict()
dictt = texas_local.to_dict()
dictc = caltech_local.to_dict()
dicth = harvard_local.to_dict()
```

```python
for attr_name, attr_value in dictb.items():
    nx.set_node_attributes(bG, attr_name, attr_value)
    
for attr_name, attr_value in dictp.items():
    nx.set_node_attributes(pG, attr_name, attr_value)
    
for attr_name, attr_value in dicto.items():
    nx.set_node_attributes(oG, attr_name, attr_value)
    
for attr_name, attr_value in dictt.items():
    nx.set_node_attributes(tG, attr_name, attr_value)
    
for attr_name, attr_value in dictc.items():
    nx.set_node_attributes(cG, attr_name, attr_value)
    
for attr_name, attr_value in dicth.items():
    nx.set_node_attributes(hG, attr_name, attr_value)
```

**Making the lists of attribute assortativity coefficients of different features:**

```python
status = [] 
gender = [] 
major = []
second_m = [] 
dorm_house = [] 
year = [] 
h_school = []

graphs = dict(zip(names,(bG, cG, hG, oG, pG, tG)))
```

```python
for graph in graphs.values():
    status.append(nx.attribute_assortativity_coefficient(graph,'student/faculty status flag'))
    gender.append(nx.attribute_assortativity_coefficient(graph,'gender'))
    major.append(nx.attribute_assortativity_coefficient(graph,'major'))
    second_m.append(nx.attribute_assortativity_coefficient(graph,'second major/minor'))
    dorm_house.append(nx.attribute_assortativity_coefficient(graph,'dorm/house'))
    year.append(nx.attribute_assortativity_coefficient(graph,'year'))
    h_school.append(nx.attribute_assortativity_coefficient(graph,'high school'))
    
```

**Finally building data frame with all assortativity coefficients of all universities:**

```python
assortativity = pd.DataFrame.from_records([status, gender, major, second_m, dorm_house, year, h_school], 
                                          columns=[list(names)])
assortativity.index = ['student/faculty status flag', 'gender', 'major', 'second major/minor', 'dorm/house', 
                         'year', 'high school']
assortativity
```

**Here we see that 'gender', 'major' and 'second major/minor' have the smaller assortativity coefficients in all universities than 'student/faculty status flag' and 'year' features. It shows that these features are more important in the mechanism of connecting people in the network. In some cases living in dorm or house (in Oklahoma and Harvard, for example) is also has an impact on connectivity.**
